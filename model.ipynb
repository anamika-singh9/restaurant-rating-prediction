Dividing detaset into (x data) input features and (y data) target and dependent variable. 

```python
x = df[['Average Cost for two', 'Has Table booking', 'Has Online delivery', 'Price range']]
```

```python
y = df['Aggregate rating']
```

```python
x
```

Data Encoding and feature engineering

```python
x["Price range"].value_counts()
```

Data encoding . 

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
x["Has Table booking"] = labelencoder.fit_transform(x["Has Table booking"])
```

```python
x
```

```python
# Has Table booking 0 -> No and 1 -> yes

labelencoder.classes_
```

```python
x["Has Online delivery"] = labelencoder.fit_transform(x["Has Online delivery"])
```

```python
x
```

```python
# Has Online delivery 0 -> No and 1 -> yes

labelencoder.classes_
```

Performing standard scaling .

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

Splitting Dataset into training and testing datasets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

creating functions for finding errors. 

```python
def modelresults(predictions):
    print("Mean abslute error on model is {}".format(mean_absolute_error(Y_test,predictions)))
    print("Root mean squared error on model is {}".format(np.sqrt(mean_squared_error(Y_test,predictions))))
```

Creating  multipole models. 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
predictionsfromlr=lr.predict(X_test)
```

```python
modelresults(predictionsfromlr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel, param_grid = param_grid)
```

```python
grid_tree.fit(X_train, Y_train)
```

```python
treepredictions = grid_tree.predict(X_test)
```

```python
modelresults(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {"max_depth" : [5,15], "n_estimators" : [2,5,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train, Y_train)
```

```python
randomforestpredictions = gridrfr.predict(X_test)
```

```python
modelresults(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
grid_tree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors": [9,11,14,15,16]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```

```python
knn_grid_search.fit(X_train, Y_train)
```

```python
knnpreds = knn_grid_search.predict(X_test)
```

```python
modelresults(knnpreds)
```

```python
knn_grid_search.best_params_
```

  Analyze the most influential features affecting restaurant ratings

```python
X = df[['Average Cost for two', 'Has Table booking', 
        'Has Online delivery', 'Price range']]

best_model = grid_tree.best_estimator_

importance = best_model.feature_importances_

feature_importance = pd.DataFrame({
    'Feature': X.columns,  
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

print(feature_importance)
```
